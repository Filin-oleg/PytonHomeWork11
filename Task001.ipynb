## Дана функция f(x) = -5*x**2 + 10*x - 150
### Задание:
1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
from sympy import *
from sympy.plotting import plot
from sympy.solvers.inequalities import solve_univariate_inequality
```

```python
x = Symbol('x')
f1 = -5*x**2 + 10*x - 150
f1
```

1. Определить корни

```python
x1, x2 = solve(f1)
print('Первый корень  = ', x1.evalf())
print('Второй корень  = ', x2.evalf())
```

```python
f2 = diff(f1)
extremum = solve(f2)
```

2. Найти интервалы, на которых функция возрастает

```python
fun1 = plot(f2, (x, -2, 0), show=False)
fun1.show()
print(f'Функция возрастает с {extremum} до +ꝏ')
```

3. Найти интервалы, на которых функция убывает

```python
fun1 = plot(f2, (x, -2, 0), show=False)
fun1.show()
print(f'Функция убывает от -ꝏ до {extremum}')
```

4. Построить график

```python
fun2 = plot(f1, (x, -11, 12), show=False)
fun2.show()
```

5. Вычислить вершину

```python
print(f'Вершина: {extremum}')
```

6. Определить промежутки, на котором f > 0

```python
print(f'Функция {f1} больше нуля от -00 до {round(x2.evalf(),2)} и от {round(x1.evalf(),2)} до +00')
```

7. Определить промежутки, на котором f < 0

```python
print(f'Функция {f1} меньше нуля от {round(x2.evalf(),2)} и до {round(x1.evalf(),2)}')
```
